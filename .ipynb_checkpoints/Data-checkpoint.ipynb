# Imputer (on final_df)

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0, missing_values=np.nan)

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
data_numeric = grouped_df.select_dtypes(include=numerics)
imputed_final_df = pd.DataFrame(imputer.fit_transform(data_numeric), columns=grouped_df.columns)
```

```python
imputed_final_df
```

# Create the control dataset

```python
conflict_years = final_df['CountryYear']
```

```python
csv4_after_1980 = csv4[csv4['Year'] > 1975]
csv4_after_1980_noconflict = csv4_after_1980[csv4_after_1980.CountryYear.isin(conflict_years)==False]
csv4_after_1980.shape[0] - csv4_after_1980_noconflict.shape[0]
control_sample = csv4_after_1980.sample(110)[['CountryYear']]
```

```python
control_sample

def slice_fn(sample):
    year = int(sample[-4:])
    words = [sample[:-4]+str(year-i-1) for i in range(4)]
    words.append(sample)
    return words

full_list = []

for index, row in control_sample.iterrows():
    full_list.append(slice_fn(row['CountryYear']))    
```

```python
from itertools import chain

samples = list(chain(*full_list))

sample_full = csv_full[csv_full.CountryYear.isin(samples)==True]
sample_full
```

```python
sample_full['conflict_id'] = 0
```

```python
id_list = list(range(111,221))*5
```

```python
id_list = sorted(id_list)
```

```python
sample_full['conflict_id'] = id_list
```

```python
control_groupedby = sample_full.groupby('conflict_id').mean()
```

```python
columns_to_keep = list(imputed_final_df.columns)
```

```python
sample_full_clean = control_groupedby.drop(columns=[col for col in control_groupedby if col not in columns_to_keep])

sample_full_clean.isna().sum()
```

# Imputing control


```python
imputed_control = pd.DataFrame(imputer.fit_transform(sample_full_clean), columns=sample_full_clean.columns)
imputed_control
```

# LET'S MODEL SOME SHT

```python
imputed_final_df['target'] = 1
imputed_control['target']=0
```

```python
full_dataset = pd.concat([imputed_control,imputed_final_df])

full_dataset
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

X = full_dataset.drop(columns='target')
y = full_dataset['target']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3)
```

```python
bst = XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic')
```

```python
bst.fit(X_train, y_train)
```

```python
bst.feature_importances_
```

```python
preds = bst.predict(X_test)

accuracy_score(preds,y_test)
```

```python
import pandas as pd

features = bst.get_booster().feature_names
importances = bst.feature_importances_
feature_importances_df = pd.DataFrame(zip(features, importances), columns=['feature', 'importance']).set_index('feature')
feature_importances_df.sort_values('importance', ascending=False).head(10)
```
