# Function list

```python
def country_info(x):
    return conflict_df['country'].iloc[x], conflict_df['start_date'].iloc[x]
```

```python
def random_country(lower_bound, upper_bound):
    random_country = random.choice(tuple(all_countries_list))
    random_5year = old_features[(old_features['Country']==random_country) & ((old_features['Year'] >= lower_bound) & (old_features['Year'] <= upper_bound))].copy()
    return random_5year
```

```python
def war_automater(x):
    country, start_date = country_info(x)
    start_date = start_date - 1
    five_years_before = start_date-4
    five_year_df = old_features[(old_features['Country']==country) & ((old_features['Year'] >= five_years_before) & (old_features['Year'] <= (start_date)))].copy()
    return five_year_df
```

```python
def pair_automater(x):
    country, start_date = country_info(x)
    start_date = start_date - 1
    five_years_before = start_date-4
    
    country_5y = old_features[(old_features['Country']==country) & ((old_features['Year'] >= five_years_before) & (old_features['Year'] <= start_date))].copy()
    rando = random_country(five_years_before, start_date)
    while len(rando) != 5:
        rando = random_country(five_years_before, start_date)
    country_5y['Conflict'] = 1
    rando['Conflict'] = 0
    return pd.concat([country_5y, rando])
```

```python
def complete_automater(x):
    df = war_automater(x)
    df = df.dropna(axis = 1)
    name = df['Country'].iloc[1]
    df = df.drop(columns = ['Country'])
    df = df.reset_index()
    df = df.drop(columns = 'index')
#     print(name)
    return df, name
```

```python
def pair_complete_automater(x):
    df = pair_automater(x)
    df = df.dropna(axis = 1)
    name = df['Country'].iloc[1], df['Country'].iloc[6]
    df = df.drop(columns = ['Country'])
    df = df.reset_index()
    df = df.drop(columns = 'index')
#     print(name)
    return df, name
```

```python
def pct(x):
    df, name = complete_automater(x)
    p_change = df.pct_change()
    p_change.drop(columns = 'Year')
    p_change['Year'] = df['Year']
    final_pct = p_change.drop(df.index[0])
    return final_pct
```

```python
def pair_pct(x):
    df, name = pair_complete_automater(x)
    df_c = df[df['Conflict'] == 1]
    df_r = df[df['Conflict'] == 0]
    p_change_c = df_c.pct_change()
    p_change_r = df_r.pct_change()
    p_change = pd.concat([p_change_c, p_change_r])
    p_change.drop(columns = 'Year')
    p_change['Year'] = df['Year']
    p_change.drop(columns = 'Conflict')
    p_change['Conflict'] = df['Conflict']
    final_pct = p_change.drop(df.index[[0, -5]])
    return final_pct
```

```python
def catplot_pct(x):
    df = pct(x)
    dfm = df.melt('Year', var_name='cols', value_name='vals')
    sns.catplot(x="Year", y="vals", hue='cols', data=dfm, kind='point', palette = 'pastel')
```

```python
def pct_mean(x):
    df, name = complete_automater(x)
    p_change = df.pct_change()
    p_change.drop(columns = 'Year')
    p_change['Country'] = name
    final_pct = p_change.drop(df.index[0])
    final_pct = final_pct.groupby('Country').mean()
    final_pct['Year'] = df['Year'].iloc[-1]
    return final_pct
```

```python
def pair_pct_mean(x):
    df, name = pair_complete_automater(x)
    df_c = df[df['Conflict'] == 1]
    df_r = df[df['Conflict'] == 0]
    p_change_c = df_c.pct_change()
    p_change_c['Country'] = name[0]
    p_change_c['Conflict'] = df_c['Conflict']
    p_change_r = df_r.pct_change()
    p_change_r['Country'] = name[1]
    p_change_r['Conflict'] = df_r['Conflict']
    p_change = pd.concat([p_change_c, p_change_r])
    p_change.drop(columns = 'Year')
    final_pct = p_change.drop(df.index[0])
    final_pct = final_pct.groupby('Country').mean()
    final_pct['Year'] = df['Year'].iloc[-1]
    return final_pct
```

```python
def all_pcts():
    full_df = pd.DataFrame()
    for i in range(0, len(conflict_df)):
        x = pct_mean(i)
        full_df = pd.concat([full_df,x])
    return full_df
```

```python
def pair_all_pcts():
    full_df = pd.DataFrame()
    for i in range(0, len(conflict_df)):
        x = pair_pct_mean(i)
        full_df = pd.concat([full_df, x])
    return full_df
```

```python
def pair_all_complete():
    full_df = pd.DataFrame()
    for i in range(0, len(conflict_df)):
        x, y = pair_complete_automater(i)
        x['Country'] = y
        full_df = pd.concat([full_df, x])
    return full_df
```
